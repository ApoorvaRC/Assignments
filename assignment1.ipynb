<b>2. Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line.</b>

```python
for i in range(2000,3201):
    # The % (modulo) operator yields the remainder from the division of the first argument by the second
    if (i%7)==0 and (i%5)!=0:
        print(i,end=',')# end=',' means comma is appended after the last value
```

<b>4.Write a Python program to find the volume of a sphere with diameter 12 cm.
Formula: V=4/3 * π * r^3</b>


```python
#Converting Diameter into radius
diameter=float(input("enter the diameter:"))
radius=float(diameter/2)#sphere radius
print("radius",radius)
pi=3.14159
Volume = (float(4/3)*(pi)*(radius)**3)

print("Volume of the sphere:",Volume)
```

2.<b>Create the below pattern using nested for loop in Python.</b>   


pattern :
#*
#* *
#* * *  
#* * * *  
#* * * * *  
#* * * * 
#* * *  
#* * 


```python
n=5;
for i in range(n):
    for j in range(i):
        print ('* ', end="")
    print('')

for i in range(n):
    for j in range(n-i):
        print('* ', end="")
    print('')
```
